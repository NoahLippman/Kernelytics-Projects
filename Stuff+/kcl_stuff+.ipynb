```python
import pandas as pd
from scipy.stats import rankdata

# Load your file (replace with your actual filename if different)
df = pd.read_csv('KCL_stuff+.csv')

# Drop rows with missing stuff+ or TaggedPitchType or Pitcher
df = df.dropna(subset=['stuff+', 'TaggedPitchType', 'Pitcher'])

# Group by Pitcher and TaggedPitchType, calculate average stuff+
grouped = df.groupby(['Pitcher', 'TaggedPitchType'])['stuff+'].mean().reset_index()

# For each TaggedPitchType, calculate percentile for each player
def add_percentile(sub_df):
    sub_df['stuff+_percentile'] = rankdata(sub_df['stuff+'], method='average') / len(sub_df) * 100
    return sub_df

result = grouped.groupby('TaggedPitchType', group_keys=False).apply(add_percentile).reset_index(drop=True)
result = result.sort_values(['TaggedPitchType', 'stuff+'], ascending=[True, False])

# Save to CSV
result.to_csv('player_stuffplus_percentiles.csv', index=False)

# Also: Group by Pitcher, calculate average stuff+ across all pitch types and sort
pitcher_avg = df.groupby('Pitcher')['stuff+'].mean().reset_index()
pitcher_avg = pitcher_avg.sort_values('stuff+', ascending=False)
pitcher_avg.to_csv('pitcher_average_stuffplus.csv', index=False)

# Display both results
print("Pitcher + Pitch Type Percentiles:")
display(result)
print("Pitcher Average Stuff+ (All Types):")
display(pitcher_avg)
```
