```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_right = pd.read_csv('right_mean_pitch_delta_run_exp_by_outcome_count.csv')
df_left = pd.read_csv('left_mean_pitch_delta_run_exp_by_outcome_count.csv')
```

```python

# Create count column (Balls-Strikes)
df_right['Count'] = df_right['Balls'].astype(str) + '-' + df_right['Strikes'].astype(str)

# Filter for standard counts only
counts_of_interest = ['0-0', '0-1', '0-2', '1-0', '1-1', '1-2', '2-1', '2-2', '3-2']
df_right = df_right[df_right['Count'].isin(counts_of_interest)]

# Pivot the table and round to 4 decimals
pivot_table = df_right.pivot_table(
    index='TaggedPitchType',
    columns='Count',
    values='delta_run_exp_mean',
    aggfunc='mean'
).fillna(0).reindex(columns=counts_of_interest).round(4)

# Plot heatmap
plt.figure(figsize=(12, len(pivot_table) * 0.5 + 2))  # Dynamic height
sns.set(font_scale=0.9)

ax = sns.heatmap(
    pivot_table,
    annot=True,
    cmap='coolwarm',
    fmt=".4f",
    linewidths=0.5,
    linecolor='gray',
    cbar_kws={'label': 'Delta Run Exp Mean'}
)

plt.title("Delta Run Exp Mean by Pitch Type and Count", fontsize=14, weight='bold', pad=10)
plt.ylabel("Pitch Type")
plt.xlabel("Count")
plt.tight_layout()
plt.show()

```
