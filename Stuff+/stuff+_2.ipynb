```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
import glob
import numpy as np
import xgboost as xgb
```

```python
cornbelters_files = glob.glob("../KCLData/*.csv")
other_files = glob.glob("../CornBeltersData/*.csv")

all_files = cornbelters_files + other_files

# Read and concatenate all CSVs into one DataFrame
stuff_plus = [pd.read_csv(f) for f in all_files]
df = pd.concat(stuff_plus, ignore_index=True)
```

```python
features = ['RelSpeed','SpinRate','RelHeight','RelSide','Extension','InducedVertBreak', 'HorzBreak']
```

```python
FF = df[df['TaggedPitchType'] == 'Fastball']
SI = df[df['TaggedPitchType'] == 'Sinker']
FC = df[df['TaggedPitchType'] == 'Cutter']
CH = df[df['TaggedPitchType'] == 'Changeup']
CB = df[df['TaggedPitchType'] == 'Curveball']
SL = df[df['TaggedPitchType'] == 'Slider']
```

```python
FF = FF[features+['PlayResult']]
SI = SI[features+['PlayResult']]
FC = FC[features+['PlayResult']]
CH = CH[features+['PlayResult']]
CB = CB[features+['PlayResult']]
SL = SL[features+['PlayResult']]
```

```python
FF = FF.dropna()
FF = SI.dropna()
FC = FC.dropna()
CH = CH.dropna()
CB = CB.dropna()
SL = SL.dropna()
```

```python
for df_pitch in [FF, SI, FC, CH, CB, SL]:
    df_pitch['PlayResult'] = df_pitch['PlayResult'].replace(
        {'StrikeoutLooking': 'Out', 'StrikeoutSwinging': 'Out'}
    )
```

```python
map_data = {
    'Out': .247,
    'Single': -.469,
    'Double': -.776,
    'Error': 0,
    'Walk': -.222,
    'Triple': -1.051,
    'HomeRun': -1.377,
    'FieldersChoice': 0,
    'Sacrifice': 0,
    'HItByPitch': -.390,
    'HitByPitch': -.390
}

FF['PlayResult'] = FF['PlayResult'].map(map_data)
SI['PlayResult'] = SI['PlayResult'].map(map_data)
FC['PlayResult'] = FC['PlayResult'].map(map_data)
CH['PlayResult'] = CH['PlayResult'].map(map_data)
CB['PlayResult'] = CB['PlayResult'].map(map_data)
SL['PlayResult'] = SL['PlayResult'].map(map_data)

FF['RelSide'] = abs(FF['RelSide'])
SI['RelSide'] = abs(SI['RelSide'])
FC['RelSide'] = abs(FC['RelSide'])
CH['RelSide'] = abs(CH['RelSide'])
CB['RelSide'] = abs(CB['RelSide'])
SL['RelSide'] = abs(SL['RelSide'])
```

```python
pitch_types = ['FF', 'SI', 'FC', 'CH', 'CB', 'SL']
models = {}
```

```python
from scipy.stats import uniform, randint
from sklearn.model_selection import KFold, RandomizedSearchCV
from math import sqrt
import pickle
# Map pitch type codes to DataFrames
pitch_type_map = {
    'FF': FF,
    'SI': SI,
    'FC': FC,
    'CH': CH,
    'CB': CB,
    'SL': SL
}
mean_playresult_dict = {}


for pitch in pitch_types:
    df_pitch = pitch_type_map[pitch].copy()
    # Drop rows with missing target or features
    df_pitch = df_pitch.dropna(subset=['PlayResult'] + features)
    X = df_pitch[features]
    y = df_pitch['PlayResult']

    if len(X) < 50:
        print(f"Not enough data for {pitch}, skipping.")
        continue

    # Define the XGBoost Regressor model
    xgb_model_rand = xgb.XGBRegressor(objective='reg:squarederror',
                                      eval_metric='rmse',
                                      random_state=42)

    # Define the parameter distributions to sample from
    param_distributions = {
        'n_estimators': randint(100, 500),
        'learning_rate': uniform(0.01, 0.2),
        'max_depth': randint(3, 10),
        'min_child_weight': randint(1, 6),
        'subsample': uniform(0.6, 0.4),
        'colsample_bytree': uniform(0.6, 0.4),
        'gamma': uniform(0, 0.5),
        'reg_alpha': uniform(0, 0.1),
        'reg_lambda': uniform(0, 1)
    }

    # Set up RandomizedSearchCV
    random_search = RandomizedSearchCV(
        estimator=xgb_model_rand,
        param_distributions=param_distributions,
        n_iter=50,
        cv=KFold(n_splits=5, shuffle=True, random_state=42),
        scoring='neg_mean_squared_error',
        verbose=2,
        random_state=42,
        n_jobs=-1
    )

    # Fit the RandomizedSearchCV object to your data
    print(f"\nStarting RandomizedSearchCV for {pitch}...")
    random_search.fit(X, y)

    # Get the best parameters and best score
    print(f"Best parameters found for {pitch}: {random_search.best_params_}")
    print(f"Best cross-validation RMSE for {pitch}: {sqrt(-random_search.best_score_):.4f}")
    print(y.mean())
    xgb_model = xgb.XGBRegressor(**random_search.best_params_)
    xgb_model.fit(X, y)
    models[pitch] = xgb_model
    mean_playresult = y.mean()
    mean_playresult_dict[pitch] = mean_playresult
    # Save model
    with open(f"models/stuff_plus_model_{pitch}.pkl", "wb") as f:
        pickle.dump(xgb_model, f)
    with open(f"models/stuff_plus_{pitch}_mean_playresult.pkl", "wb") as f:
        pickle.dump(mean_playresult_dict, f)
```

```python
def predict_stuff_plus(event):
    # Make sure all required features are present
    features = ['RelSpeed', 'SpinRate', 'RelHeight', 'RelSide', 'Extension', 'InducedVertBreak', 'HorzBreak']
    if all(f in event and not pd.isna(event[f]) for f in features):
        # Predict using the trained model
        pred = xgb_model.predict([[event[f] for f in features]])[0]
        # Scale so mean is 100 (use your training mean or pass as argument if needed)
        mean_playresult = y.mean()  # or use a fixed value if you want
        return pred * (100 / mean_playresult)
    else:
        return np.nan  # Not enough data to predict
```

```python
# event should be a dict or Series with the 7 features
import warnings
warnings.filterwarnings("ignore")
kcl_df['stuff+'] = kcl_df.apply(predict_stuff_plus, axis = 1)
cornbelters_df['stuff+'] = cornbelters_df.apply(predict_stuff_plus, axis = 1)
kcl_df.to_csv("kcl_stuff+_data.csv", index=False)
cornbelters_df.to_csv("cornbelters_stuff+_data.csv", index=False)
```
