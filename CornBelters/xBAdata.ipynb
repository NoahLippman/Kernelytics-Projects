```python
ab_results = [
    'Out', 'StrikeoutLooking', 'Single', 'StrikeoutSwinging',
    'Double', 'HomeRun', 'Error'
]

# Calculate at-bats for each individual player in the Batter column
ab_counts = df[df['PlayResult'].isin(ab_results)].groupby('Batter').size().reset_index(name='at_bats')
ab_counts
```

```python
# Calculate total xBA for each batter
xba_sum = df[df['PlayResult'].isin(ab_results)].groupby('Batter')['xBA'].sum().reset_index(name='total_xBA')

# Merge with ab_counts to get at_bats and total_xBA together
batter_stats = pd.merge(ab_counts, xba_sum, on='Batter', how='left')

# Calculate xBA per at-bat for each batter
batter_stats['xBA'] = batter_stats['total_xBA'] / batter_stats['at_bats']

# Keep only Batter and xBA_per_at_bat columns
batter_xba = batter_stats[['Batter', 'xBA']]
batter_xba
```

```python
batter_xba.to_csv('Data/xBA_Data.csv', index=False)
```
