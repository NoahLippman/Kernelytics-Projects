```python
import numpy as np
from numpy import random
```

Simulates 1,000,000 9th innings & loops through P-values to find the p-value that gives us a simulated walkoff probability of 20%

```python
def simInning (pVal):
    batterProbs = {1 : [.6, .3, .1], 2 : [.6, .3, .1], 3 : [.6, .25, .15], 
                   4 : [.9 - pVal, pVal, .1], 5 : [.9 - pVal, pVal, .1], 6 : [.6, .25, .15], 
                   7 : [.6, .3, .1], 8 : [.6, .3, .1], 9 : [.6, .3, .1]}
    
    outcomes = []
    for _ in range(1000000):
        baseState = {1: False, 2: False, 3: False}
        outs = 0
        runs = 0
        batter = 1
        while (outs < 3 and runs < 3):
            randomNum = random.random()
            batterProb = batterProbs.get(batter)
            if randomNum < batterProb[0]:
                outs += 1
            elif randomNum < batterProb[0] + batterProb[1]:
                if (baseState.get(3) == True):
                    baseState[3] = False
                    runs += 1
                if (baseState.get(2) == True):
                    baseState[2] = False
                    runs += 1
                if (baseState.get(1) == True):
                    baseState[3] = True
                baseState[1] = True
            else:
                runs = runs + baseState.get(1) + baseState.get(2) + baseState.get(3) + 1
                baseState[1] = 0
                baseState[2] = 0
                baseState[3] = 0
            batter += 1

        outcomes.append(runs > 2)
    return(np.mean(outcomes))

for p in np.arange(0,.5,.01):
    print([p, simInning(p)])
```

Based on 1,000,000 simulations, the value of p is approximately equal to .38 when the home team has a 20% overall chance to walk off and win the game in the 9th inning
